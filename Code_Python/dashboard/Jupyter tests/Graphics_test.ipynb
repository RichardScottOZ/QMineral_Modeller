```python
import pandas as pd
```

```python
#df_new = pd.read_csv('evandro_small.csv',encoding='latin-1')

df_new = pd.read_csv('teste.csv',encoding='latin-1')
df_new = df_new.loc[:, ~df_new.columns.str.contains('^Unnamed')]
df_new
```

```python
import numpy as np

features = df_new.columns.to_list()
clean_features = []

for name in features:
    if df_new[name].dtypes == np.float or df_new[name].dtypes == np.int64:
        clean_features.append(name)
clean_features
```

```python
import plotly.graph_objects as go


def select_axis(df,feature,axis):
  
    return dict(
        args=[ {axis:[df[feature]], 
                'name':feature}],
        label=feature,
        method="update")
    

fig = go.Figure()

# Add traces
fig.add_trace(
    go.Scatter(
        x=df_new[clean_features[0]],
        y=df_new[clean_features[0]],
        mode="markers",
        marker=dict(color='rgba(152, 0, 0, .8)')
    )
)

###### if want to add axis labels ######
#fig.update_layout(
#    xaxis_title=clean_features[0],
#    yaxis_title=clean_features[0],
#                  )

# Add dropdown

fig.update_traces(mode='markers', marker_line_width=2, marker_size=10)

###### if want to add title ######
#fig.update_layout(
#    title={
#        'text': "Biplot",
#        'y':0.9,
#        'x':0.5,
#        'xanchor': 'center',
#        'yanchor': 'top'})

fig.update_layout(
    updatemenus=[
        dict(
            buttons=list(
                    [select_axis(df_new,feature,'x') for feature in clean_features],
            ),
    
            direction="down",
            pad={"r": 10, "t": 10},
            showactive=True,
            x=-0.05,
            xanchor="left",
            y=1.15,
            yanchor="top"
        ),
        
        dict(
            buttons=list(
                    [select_axis(df_new,feature,'y') for feature in clean_features],
            ),
            
     
            direction="up",
            pad={"r": 30, "t": 50},
            showactive=True,
            x=1.05,
            xanchor="right",
            y=-0.05,
            yanchor="middle"
        )]        
             )


fig.show()
```

```python
import plotly.graph_objects as go

########## TODO ###########
# Color by attribute Group and Minerals
# Change the hoever style
colors = [
    '#1f77b4',  #// muted blue
    '#ff7f0e',  #// safety orange
    '#2ca02c',  #// cooked asparagus green
    '#d62728',  #// brick red
    '#9467bd',  #// muted purple
    '#8c564b',  #// chestnut brown
    '#e377c2',  #// raspberry yogurt pink
    '#7f7f7f',  #// middle gray
    '#bcbd22',  #// curry yellow-green
    '#17becf'   #// blue-teal
    '#800000'   #// Olive
    '#008000'   #// Green
    '#800080'   #// Purple
    '#008080'   #// Teal
    '#000080'   #// Navy
]



def select_axis(df,feature,axis):
  
    return dict(
        args=[ {axis:[df[feature]]}], 
                #'name':feature}],
        label=feature,
        method="update")

def makeAxis(title, tickangle):
    return {
      'title': title,
      'titlefont': { 'size': 20 },
      'tickangle': tickangle,
      'tickfont': { 'size': 12 },
      'tickcolor': 'rgba(0,0,0,0)',
      'ticklen': 5,
      'showline': True,
      'showgrid': True
    }
    
# g = list(df_new['GROUP'].unique())
# g = list(df_new['MINERAL'].unique())
# k = 0

fig = go.Figure()

fig.update_layout({
    'ternary': {
        'sum': 100,
        'aaxis': makeAxis('', 0),
        'baxis': makeAxis('', 45),
        'caxis': makeAxis('', -45)
    }
})

fig.update_layout(
    updatemenus=[
        dict(
            buttons=list(
                    [select_axis(df_new,feature,'a') for feature in clean_features],
            ),
    
            direction="down",
            pad={"r": 10, "t": 10},
            showactive=True,
            x=0.45,
            xanchor="left",
            y=1.20,
            yanchor="top"
        ),
        
        dict(
            buttons=list(
                    [select_axis(df_new,feature,'b') for feature in clean_features],
            ),
            
     
            direction="up",
            pad={"r": 30, "t": 50},
            showactive=True,
            x=0.32,
            xanchor="right",
            y=-0.07,
            yanchor="middle"
        ),

        dict(
            buttons=list(
                    [select_axis(df_new,feature,'c') for feature in clean_features],
            ),
            
     
            direction="up",
            pad={"r": 30, "t": 50},
            showactive=True,
            x=0.85,
            xanchor="right",
            y=-0.07,
            yanchor="middle"
        ),
        
       dict(
            buttons=list([
                dict(
                    args=["reversescale", False],
                    label="Group",
                    method="update"
                ),
                dict(
                    args=["reversescale", True],
                    label="Mineral",
                    method="update"
                )
            ]),
            type = "buttons",
            direction="right",
            pad={"r": 10, "t": 10},
            showactive=True,
            x=0.75,
            xanchor="left",
            y=1.065,
            yanchor="top"
        )
    ])

a = fig['layout']
b = a['updatemenus'][-1]
for option in b['buttons']:
    if option['args'][-1] == True:
        select = option['label'].upper()

g = list(df_new[select].unique())
k = 0

for i in g:
    df_plot = df_new[df_new[select] == i]
    texthover = '<b>%{hovertext}</b><br><br>'+select + '=' + i + '<br>a=%{a}<br>b=%{b}<br>c=%{c}<extra></extra>'
    fig.add_trace(go.Scatterternary({
    'mode': 'markers',
    'a': df_plot[clean_features[0]],
    'b': df_plot[clean_features[0]],
    'c': df_plot[clean_features[0]],
    'text': df_plot.index,
    'legendgroup': i,
    'hovertemplate': texthover,
    'hovertext': df_plot[select],
    'name': i,
    'marker': {
        'symbol': "circle",
        'color': colors[k],        
        'size': 8,
        'line': { 'width': 2 }
    }
    }))
    k += 1


fig.update_layout(
    annotations=[
        dict(text="Group by", x=0.86, xref="paper", y=1.125, yref="paper",
                             showarrow=False)
    ])


fig.show()
```
